```python
import time
from kafka import KafkaProducer
from kafka import KafkaConsumer
```

```python
kafka_url = 'localhost:9092'
```

```python
producer_1 = KafkaProducer(
    bootstrap_servers=kafka_url,
    value_serializer=lambda v: str(v).encode('utf-8')
)
```

```python
topic = 'exo1'
```

```python
for i in range(20):
    message = f'Message {i}'
    producer_1.send(topic, message)
    print(f"Message sent. {message}")
    time.sleep(1)

producer_1.flush()
producer_1.close()
```

```python
consumer_1 = KafkaConsumer(
    topic,
    bootstrap_servers=[kafka_url],
    auto_offset_reset='earliest',
    consumer_timeout_ms=1000,
    value_deserializer=lambda v: v.decode('utf-8'),
)
```

```python
for msg in consumer_1:
    print(msg.value)

consumer_1.close()
```

```python
temperatures = []
humidity = []
press = []
```

```python
max_len = len(temperatures)

if max_len > len(humidity):
    max_len = len(humidity)

if max_len > len(press):
    max_len = len(press)

meteo_vals = []

for i in range(max_len):
    meteo_vals.append({
        "temperature": temperatures[i],
        "humidity": humidity[i],
        "press": press[i]
    })
```

```python
import json

producer_3 = KafkaProducer(
    bootstrap_servers=kafka_url,
    value_serializer=lambda v: str(v).encode('utf-8')
)

topic = 'meteo'

for val in meteo_vals:
    producer_3.send(topic, json.dumps(val))
    print(f"Message sent. {json.dumps(val)}")
    time.sleep(1)

producer_3.flush()
producer_3.close()
```

```python
consumer_1 = KafkaConsumer(
    topic,
    bootstrap_servers=[kafka_url],
    auto_offset_reset='earliest',
    consumer_timeout_ms=1000,
    value_deserializer=lambda v: v.decode('utf-8'),
)

temp_tot = 0
humid_tot = 0
press_tot = 0

for msg in consumer_1:
    values = json.loads(msg.value)
    temp_tot += int(values['temperature'])
    humid_tot += int(values['humidity'])
    press_tot += int(values['press'])

day_avg = {
    "temperature": temp_tot / max_len,
    "humidity": humid_tot / max_len,
    "press": press_tot / max_len,
}

consumer_1.close()
```

```python
import json

producer_3 = KafkaProducer(
    bootstrap_servers=kafka_url,
    value_serializer=lambda v: str(v).encode('utf-8')
)

topic = 'day_avg'

producer_3.send(topic, json.dumps(day_avg))
print(f"Message sent. {json.dumps(day_avg)}")

producer_3.flush()
producer_3.close()
```
